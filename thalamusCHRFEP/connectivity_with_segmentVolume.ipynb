```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import os
import re
import scipy
from scipy import stats
from statsmodels.stats.anova import anova_lm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix
```

```python
allData=pd.read_csv("/Users/admin/Dropbox/CHRvsNOR/2014_stats/37_38_22_thr10/LR_averaged_thalData.csv",sep='\t')
```

```python
chrData=allData[allData.group=='CHR']
fepData=allData[allData.group=='FEP']
norData=allData[allData.group=='CON']
```

##Correlation between total connectivity from thalamus to cortex with the segment volume

```python
for CORTEX in chrData.cortex.unique():
    num=0
    for groupData in norData,chrData,fepData:
        if num==0:
            printGroup='CHR'
        elif num==1:
            printGroup='FEP'
        elif num==2:
            printGroup='CON'
            
        df = groupData[groupData.cortex==CORTEX]
        coeff,pVal=scipy.stats.pearsonr(df['totalConnectivity'],df['segmentVolume'])
        
        print '\t',printGroup,CORTEX
        print '\t ============='
        print '\tp-value : {0}\n\tcorrelation coefficient : {1}'.format(pVal,coeff)
        plt.figure(figsize=(5,5))
        plt.scatter(df['totalConnectivity'],df['segmentVolume'],alpha=.5)


        data = pd.DataFrame({'x':df['totalConnectivity'], 'y':df['segmentVolume']})
        model = ols("y ~ x",data).fit()
        offset, coef = model._results.params

        plt.plot(df['totalConnectivity'], df['totalConnectivity']*coef + offset)
        plt.xlabel('{0} Total Connectivity'.format(CORTEX))
        plt.ylabel('segmentVolume')
        plt.show()
        print '\n\n\n'
        num=num+1

            
```
