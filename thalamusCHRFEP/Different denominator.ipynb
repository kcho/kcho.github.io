### Extracting total connectivity to each cortex as separate table

````python
df = allData.pivot(index='subject',columns='cortex',values='totalConnectivity')
````

### Divide each total connectivity by the total connectivity of somatomotor cortex

````python
denominator = 'OCC'
new=pd.DataFrame()
for column in df:
    a = df[column]/df[denominator].values
    new=pd.concat([new,pd.DataFrame(a)],axis=1)
````

### Reset the index and re-name the columns

````python
new= new.reset_index()
new.columns=[u'subject', u'LPFC', u'LTC', u'MPFC', u'MTC', u'OCC', u'OFC', u'PC', u'SMC']
````

### Add other information from allData such as 'group'

````python
allNew = new.merge(allData[allData['cortex']=='LPFC'],
                   on='subject')
````

### Statistics

````python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm


for CORTEX in [u'LPFC', u'LTC', u'MPFC', u'MTC', u'OCC', u'OFC', u'PC', u'SMC']:
    RC = allNew[CORTEX]
    GROUP = allNew['group']
#    IC = allData[allData.cortex==CORTEX].intracranialVolume

    formula = 'RC ~ GROUP'
#    formula = 'RC ~ GROUP'
    lm = ols(formula, allNew).fit()

    output=anova_lm(lm)
    if output.ix['GROUP','PR(>F)'] < 1:
        print '{0} \t\t {1}~GROUP'.format(CORTEX,CORTEX)
        print '========='
        print output
        print '\n\n\n'
    #print lm.summary()
````
