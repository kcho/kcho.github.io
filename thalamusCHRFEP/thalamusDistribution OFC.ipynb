#Absolute values

##Box plot of total connectivity in three groups

```python
df = allData[allData.cortex=='OFC'].ix[:,['subject','totalConnectivity','group']].set_index('subject')
df.boxplot(by='group', figsize=(18,18))
#allData[allData.cortex=='OFC'].ix[:,['subject','totalConnectivity','group']].set_index('subject')
```

##Total connectivity : Density plots

```python
size = 18


df=allData[allData.cortex=='OFC']
fig, axes = plt.subplots(nrows=2, ncols=2)

df[df.group=='CON']['totalConnectivity'].plot(ax=axes[1,1],figsize=(size,size), kind='kde',label='NOR',linewidth=4).set_title('All Groups')
df[df.group=='CHR']['totalConnectivity'].plot(ax=axes[1,1],kind='kde',style='orange',label='CHR',linewidth=4)
df[df.group=='FEP']['totalConnectivity'].plot(ax=axes[1,1],kind='kde',style='red',label='FEP',linewidth=4)
plt.legend()

df[df.group=='CON']['totalConnectivity'].plot(ax=axes[0,0],kind='kde',linewidth=4).set_title('NOR')
df[df.group=='CHR']['totalConnectivity'].plot(ax=axes[0,1],kind='kde',linewidth=4).set_title('CHR')
df[df.group=='FEP']['totalConnectivity'].plot(ax=axes[1,0],kind='kde',linewidth=4).set_title('FEP')
```

##Total connectivity : Histogram

```python
cutOffs=18
fig, axes = plt.subplots(nrows=2, ncols=2)

allData[allData.cortex=='OFC']['totalConnectivity'].hist(ax=axes[0,0],color='k', figsize=(18,18),alpha=0.7,bins=cutOffs).set_title('all')
allData[allData.cortex=='OFC'][allData.group=='CON']['totalConnectivity'].hist(ax=axes[0,1],color='blue',figsize=(18,18),alpha=0.7,bins=cutOffs).set_title('NOR')
allData[allData.cortex=='OFC'][allData.group=='CHR']['totalConnectivity'].hist(ax=axes[1,0],color='orange',figsize=(18,18),alpha=0.7,bins=cutOffs).set_title('CHR')
allData[allData.cortex=='OFC'][allData.group=='FEP']['totalConnectivity'].hist(ax=axes[1,1],color='red',figsize=(18,18),alpha=0.7,bins=cutOffs).set_title('FEP')
```
